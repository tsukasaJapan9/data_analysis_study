```python
from scipy import interpolate

x = np.linspace(0, 10, num=11, endpoint=True)
y = np.cos(-x**2/5.0)
f = interpolate.interp1d(x, y, "linear")
# plt.plot(x, f(y), "-")
# plt.grid(True)
```

```python
x = np.linspace(0, 10, num=11, endpoint=True)
y = np.sin(x**2/5.0)
plt.plot(x, y, "o")
plt.grid(True)
```

```python
f2 = interpolate.interp1d(x, y, "quadratic")
f3 = interpolate.interp1d(x, y, "cubic")
x_new = np.linspace(0, 10, num=30, endpoint=True)
plt.plot(x, y, "o", x_new, f2(x_new), "-", x_new, f3(x_new), "--")
plt.grid(True)
```

```python
from scipy.optimize import fsolve

def f(x):
    return 5 * x - 10
x = np.linspace(-10, 10)

plt.plot(x, f(x))
x = fsolve(f, 0)
x
```

```python
from scipy.optimize import fsolve

def f(x):
    return x**3 - 2*(x**2) - 11*x + 12
x = np.linspace(-5, 5)

plt.plot(x, f(x))
plt.plot(x, np.zeros(len(x)))
plt.grid(True)
x = fsolve(f, 0)
x
```
