```python
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0) 

# グラフの大きさ指定
plt.figure(figsize = (15, 6))

# グラフを表示
plt.pie(sizes, explode = explode, labels = labels, colors = colors,
        autopct = '%1.1f%%', startangle = 90)

# 円を丸く描画
plt.axis('equal')
```

```python
data = pd.read_csv('./chap7/dow_jones_index.data',sep=',')
data.date.head(100)
```

```python
data.info()
```

```python
data.date = pd.to_datetime(data.date)
```

```python
data
```

```python
delete_dol = lambda x: str(x).replace("$", "")

data.open = pd.to_numeric(data.open.map(delete_dol))
data.high = pd.to_numeric(data.high.map(delete_dol))
data.low = pd.to_numeric(data.low.map(delete_dol))
data.close = pd.to_numeric(data.close.map(delete_dol))
data.next_weeks_open = pd.to_numeric(data.next_weeks_open.map(delete_dol))
data.next_weeks_close = pd.to_numeric(data.next_weeks_close.map(delete_dol))
```
